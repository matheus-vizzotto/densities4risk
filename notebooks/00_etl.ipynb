Data treatment for experiment.

```python
import pandas as pd

# custom libs
import sys
sys.path.append("..")
from src.libs.data_wrangling import add_return_cols, ts_to_df
```

```python
# Dados
data_path = "../data/processed/"
returns_path = ''.join([data_path, 'ibovespa.xlsx'])
df = pd.read_excel(returns_path).loc[:, ["datetime", "value"]]
df.head()
```

```python
# mantém apenas dados em horário de negociação
df = df.set_index("datetime").between_time("10:00", "17:00").reset_index()

# mantém apenas dados no intervalo de 5 minutos
df = df[df.datetime.dt.minute % 5 == 0]

# adiciona coluna de retornos
df_returns = add_return_cols(df=df, y="value")

# Cria coluna para cada dia
df_returns_wide = ts_to_df(
                        df_returns, 
                        y='r_t', 
                        datetime_col='datetime')
df_returns_wide.head()
```
